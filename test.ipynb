# Convex functions

**Definition**: A function $f:R^n \rarr R$ is convex if its domain is a convex set and for all u, v in its domain and $\forall \lambda \in [0,1]$, we have

$$ f(\lambda u + (1-\lambda)v) \leq \lambda f(u) + (1-\lambda)f(v) $$

References

1. https://www.princeton.edu/~aaa/Public/Teaching/ORF523/S16/ORF523_S16_Lec7_gh.pdf


Visualization: 

```python
def parabola(x):
    return x**2

def sin(x):
    return np.sin(x)

def visualize_convexity(ax, points):
    ys = [parabola(x) for x in points]
    ax.scatter(points, ys, color='red', label='Random Points')
    ax.legend()
```

```python
# create domain and range
dom = np.arange(0, 10)
range_values = [parabola(x) for x in dom]

# create random points
points = np.random.randint(0, 11, 2)

fig, ax = plt.subplots()
ax.plot(dom, range_values, label="x^2")

visualize_convexity(ax, points)
ax.legend()
plt.show()
```

```python
dom = np.arange(0,10)
range = [parabola(x) for x in dom]
points = np.random.randint(0,11,2)

fig = plt.figure()
plt.plot(dom, range, label="x^2")

visualize_convexity(fig, points)
plt.legend()
plt.show()
```
